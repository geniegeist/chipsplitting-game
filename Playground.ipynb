```python
%matplotlib ipympl
import numpy as np
import matplotlib.pyplot as plt
```

```python
ax = plt.axes(projection="3d")
x_data = np.linspace(0,1,100)
y_data = np.linspace(0,1,100)
X,Y = np.meshgrid(x_data,y_data)
raw = [ 1 - x - y for (x,y) in zip(X,Y)]
Z = 1 - X - Y
Z[Z < 0]  = np.nan
ax.set_xlim(0,1)
ax.set_ylim(0,1)
ax.set_zlim(0,1)
ax.plot_surface(X,Y,Z,alpha=0.6,cmap='gnuplot')

T = np.linspace(0,1,100)
X_2 = np.power(T, 2)
Y_2 = 2 * T * (1 - T)
Z_2 = np.power(1 - T, 2)

ax.plot(X_2,Y_2,Z_2, label='Bin(2,t)', color='red')
ax.plot(Y_2,X_2,Z_2, label='Bin(2,t)', color='blue')

plt.show()
```

```python
ax2d = plt.axes()
X = np.linspace(0,1,100)
Y = np.array([ 1 - x for x in X])
Y[Y < 0]  = np.nan
ax2d.set_xlim(0,1)
ax2d.set_ylim(0,1)
ax2d.plot(X,Y,alpha=0.6)
plt.show()

# Plot statistical model
T = np.linspace(0,1,100)
X = 0.7 * T
Y = 1 - 0.7 * T
ax2d.plot(X,Y, color="red")

lambd = 0.9
X_img_red = lambd * X
Y_img_red = Y
Z_img_red = (1 - lambd) * X
```
